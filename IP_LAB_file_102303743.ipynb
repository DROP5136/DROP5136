```python
import cv2
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
#Q1 grayscale to binary(mean intensity image as threshold)
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
threshold_value=np.mean(gray)
_,binary=cv2.threshold(gray,threshold_value,255,cv2.THRESH_BINARY)
cv2.imwrite('binary image.jpg',binary)

plt.subplot(1,2,1)
plt.title('grayscale image')
plt.imshow(gray,cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f'binary image(Threshold={threshold_value:.2f})')
plt.imshow(binary,cmap='gray')
plt.axis('off')
```

```python
#Q1b grayscale to binary(inpute threshold value)
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
threshold_value=float(input("Enter threshold value(0-255)"))
_,binary=cv2.threshold(gray,threshold_value,255,cv2.THRESH_BINARY)
cv2.imwrite('binary image.jpg',binary)

plt.subplot(1,2,1)
plt.title('grayscale image')
plt.imshow(gray,cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title(f'binary image(Threshold={threshold_value:.2f})')
plt.imshow(binary,cmap='gray')
plt.axis('off')


```

```python
#Q2a rgb image to grayscale image (mean avg of three planes)
img=cv2.imread('bgrimg.jpg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray = np.mean(img, axis=2).astype(np.uint8)

plt.subplot(1,2,1)
plt.title('rgb image')
plt.imshow(img)
plt.axis('off')

plt.subplot(1,2,2)
plt.title('grayscale image')
plt.imshow(gray,cmap='gray')
plt.axis('off')

```

```python
#Q2b rgb to grayscale(input weightage of three planes (0-1), sum should be 1)

img = cv2.imread('bgrimg.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

wr = float(input("Enter weight for Red (01): "))
wg = float(input("Enter weight for Green (01): "))
wb = float(input("Enter weight for Blue (01): "))

if abs(wr + wg + wb - 1.0) > 1e-6:
    raise ValueError("The sum of all weights must be 1. Please re-enter values.")

gray = (wr * img[:, :, 0] + wg * img[:, :, 1] + wb * img[:, :, 2]).astype(np.uint8)

plt.subplot(1, 2, 1)
plt.title('Original RGB Image')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Grayscale (R:{wr}, G:{wg}, B:{wb})')
plt.imshow(gray, cmap='gray')
plt.axis('off')

plt.show()
```

```python
#Q3 border around i/p image
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('bgrimg.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

border_width = int(input("Enter border width (in pixels): "))

print("Enter border color values between 0255:")
r = int(input("Red: "))
g = int(input("Green: "))
b = int(input("Blue: "))


bordered = cv2.copyMakeBorder(
    img,
    top=border_width,
    bottom=border_width,
    left=border_width,
    right=border_width,
    borderType=cv2.BORDER_CONSTANT,
    value=[r, g, b]
)


plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(img)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f'Bordered Image ({border_width}px, color=({r},{g},{b}))')
plt.imshow(bordered)
plt.axis('off')

plt.show()
```

```python
#q5 log transformation for grayscale


gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
c=1
#s=clog(1+r),then output is scaled back to 0-255
gray_float=gray.astype(np.float32)
enhanced=c*np.log(1+gray_float)
enhanced=cv2.normalize(enhanced,None,0,255,cv2.NORM_MINMAX)
enhanced=enhanced.astype(np.uint8)

plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("log transformed")
plt.imshow(enhanced,cmap='gray')
plt.axis("off")
```

```python
#q5 log transformation for binary img


gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
threshold_value=np.mean(gray)
_,bin=cv2.threshold(gray,threshold_value,255,cv2.THRESH_BINARY)
cv2.imwrite('binaryimg.jpg',bin)
c=1
#s=clog(1+r),then output is scaled back to 0-255
bin_float=bin.astype(np.float32)
enhanced=c*np.log(1+bin_float)
enhanced=cv2.normalize(enhanced,None,0,255,cv2.NORM_MINMAX)
enhanced=enhanced.astype(np.uint8)

plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(bin,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("log transformed")
plt.imshow(enhanced,cmap='gray')
plt.axis("off")
```

```python
#q6 power law/gamma transformation
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
c=1
gamma=2
#s=c.r^gamma
gray_float=gray.astype(np.float32)/255.0
enhanced=c*(gray_float**gamma)
enhanced=np.uint8(enhanced*255)

plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("power transformed")
plt.imshow(enhanced,cmap='gray')
plt.axis("off")
```

```python
# q7 contrast stretching/ index level slicing
#s=(r-rmin)/(rmax-rmin)*(smax-smin)+smin
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
gray_float=gray.astype(np.float32)

rmin=int(input('enter min value(0-255):'))
rmax=int(input('enter max value(0-255):'))

smin=int(input('enter desired min value(0-255):'))
smax=int(input('enter desired max value(0-255):'))

#During contrast stretching, the formula can produce values like: negative values, numbers greater than 255 ,Both are invalid pixel values, so clipping fixes them.
enhanced=((gray_float-rmin)/(rmax-rmin)*(smax-smin))+smin
enhanced=np.clip(enhanced,0,255)
enhanced=enhanced.astype(np.uint8)

plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("contrast stretching")
plt.imshow(enhanced,cmap='gray')
plt.axis("off")
```

```python
# q8 histogram equalization
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
hist=cv2.equalizeHist(gray)

plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("histogram equalized")
plt.imshow(hist,cmap='gray')
plt.axis("off")
```

```python
#q10 averaging filter smoothing
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
kernelsize=5
img=cv2.blur(gray,(kernelsize,kernelsize))
plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("smooth img")
plt.imshow(img,cmap='gray')
plt.axis("off")
```

```python
#gaussian filter
#q10 averaging filter smoothing
gray=cv2.imread('grayscale.png',cv2.IMREAD_GRAYSCALE)
kernelsize=5
sigma=1.0
img=cv2.GaussianBlur(gray,(kernelsize,kernelsize),sigma)
plt.subplot(1, 2, 1)
plt.title("grayscale")
plt.imshow(gray,cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("smooth img")
plt.imshow(img,cmap='gray')
plt.axis("off")
```

```python
#weighted filter
gray = cv2.imread('grayscale.png', cv2.IMREAD_GRAYSCALE)

# weighted kernel defined by h(x,y) = max(|x|, |y|)
kernel = np.array([[1, 1, 1],
                   [1, 0, 1],
                   [1, 1, 1]], dtype=np.float32)

kernel = kernel / kernel.sum()
weighted_smooth = cv2.filter2D(gray, -1, kernel)

plt.subplot(1, 2, 1)
plt.title("original")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("weighted smoothing")
plt.imshow(weighted_smooth, cmap='gray')
plt.axis("off")

plt.show()
```
